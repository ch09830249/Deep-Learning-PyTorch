```python
import torch
with open('1342-0.txt', encoding='utf8') as f:
    text = f.read()
```

```python
lines = text.split('\n')
line = lines[200]
line
```

```python
letter_t = torch.zeros(len(line), 128) #len(line) 字元數   128:ASCII碼字元數
letter_t.shape
```

```python
lines
```

```python
for i, letter in enumerate(line.lower().strip()):    #依序讀該列文字字元  轉小寫     strip(): 移除前後特定字元
    letter_index = ord(letter) if ord(letter) < 128 else 0   #ord: 取得ASCII碼
    letter_t[i][letter_index] = 1   # 對應到的位置設為1
```

```python
def clean_words(input_str):
    punctuation = '.,;:"!?”“_-'
    word_list = input_str.lower().replace('\n',' ').split()   #replace: 將'\n'換成' '
    word_list = [word.strip(punctuation) for word in word_list]   #將前後的標點符號去掉
    return word_list

words_in_line = clean_words(line)
line, words_in_line
```

```python
word_list = sorted(set(clean_words(text)))
word2index_dict = {word: i for (i, word) in enumerate(word_list)}    #word為index i為value

len(word2index_dict), word2index_dict['impossible']  # 文本中單字總數: 7261      3394: impossible的編碼
```

```python
word_t = torch.zeros(len(words_in_line), len(word2index_dict))     # 11個words   7261種words in the text
for i, word in enumerate(words_in_line):
    word_index = word2index_dict[word]
    word_t[i][word_index] = 1
    print('{:2} {:4} {}'.format(i, word_index, word))
    
print(word_t.shape)
```
